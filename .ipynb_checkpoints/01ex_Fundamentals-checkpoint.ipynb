1\. Write the following as a list comprehension

````python
# 1
ans = []
for i in range(3):
    for j in range(4):
        ans.append((i, j))
print (ans)

print("With list comprehension")
print ([(i,j) for i in range(3) for j in range(4)]) #si puo fare doppio for

# 2
ans = map(lambda x: x*x, filter(lambda x: x%2 == 0, range(5)))
print (list(ans))

print("With list comprehension")
print ([x*x for x in range(5) if (x%2 == 0)]) #si puo fare for e if
````

2\. Convert the following function into a pure function with no global variables or side effects

````python
x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!

print("Solution")

import copy

alist = [1,2,3]
def f(alist, x):
    alist_copy = copy.copy(alist)
    for i in range(x):
        alist_copy.append(i)
    return alist_copy
    
ans = f(alist, 5)
print(ans, alist)
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def my_decorator(func):
    def wrapper():
        print("Hello! I'm Kevin")
        func()
        print("See you soon! Kevin.")
    return wrapper

@my_decorator
def cube_of_2():
    x = 2
    print(x*x*x)

cube_of_2() # -> wrapped function
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
# a) factorial with recursion

def factorial (n):

    if (n == 0):
        return 1
    else:
        return n * factorial(n-1)

print("Recursive function n! =", factorial(5))

# a) factorial without recursion

def factorial (n):

    if (n == 0):
        return 1
    else:
        fact = 1
        for i in range(0,n,1):
            fact = fact*(n-i)
        return fact

print("Without recursive function n! =", factorial(5))
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
import math as m

densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radius = [1,2,3]

for mat in densities:
    for r, dens in zip(radius, densities.values()): #.values() gives the value, .items() gives keys+values
        print(f"Radius = {r}, Element = {mat}")
        measure = [(lambda r: 2 * m.pi * r)(r), (lambda r: m.pi * pow(r,2))(r), (lambda r: 4 / 3 * m.pi * pow(r,3))(r)]
        print("measures =",measure )
        masses = [measure[i] * dens[i] for i in range(3)]
        print("masses =" ,masses)
````

````python
# Parent class
class dog:

    # Class attribute
    species = 'mammals'
    sound = "wof"
    count = 0

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.species = "mammals"
        self.is_hungry = True

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)   
        
    # instance method
    def eat(self):
        self.is_hungry = False

dog1 = dog("Tom", 6)
dog2 = dog("Fletcher", 7)
dog3 = dog("Larry", 9)

#Be carefull! Methods should be called with ".method()". Attributes only with ".attribute" !!!

print(f"I have 3 dogs. {dog1.description()}. {dog2.description()}. {dog3.description()}. And they're all {dog.species}, of course. My dogs are hungry.")
dog1.eat()
dog2.eat()
dog3.eat()  
if (not dog1.is_hungry and not dog2.is_hungry and not dog3.is_hungry):
    print(f"I have 3 dogs. {dog1.description()}. {dog2.description()}. {dog3.description()}. And they're all {dog.species}, of course. My dogs are NOT hungry.")
````
