### Ex 1

1\.a Make a new directory called `students` in your home. Download a csv file with the list of students of this lab from [here](https://www.dropbox.com/s/867rtx3az6e9gm8/LCP_22-23_students.csv) (use the `wget` command) and copy that to `students`. First check whether the file is already there

```python
# Go to home directory
cd $HOME

# Make a folder called 'students'
cd students

# Download the CSV
wget https://www.dropbox.com/s/867rtx3az6e9gm8/LCP_22-23_students.csv

# Find the file and store the path
path=$(find $HOME -name 'LCP_22-23_students.csv' -print -quit) # variabile percorso del file

# Copy it to the current directory (students)
cp "$path" .

# List the contents
ls

# Confirm it's here
find . -name 'LCP_22-23_students.csv'
```

1\.b Make two new files, one containing the students belonging to PoD, the other to Physics.

```python
touch PoD_students.csv Physics_students.csv

# echo "# Cognome,Nome" > PoD_students.csv #stampa header
#echo "# Cognome,Nome" > Physics_students.csv

grep "PoD" LCP_22-23_students.csv | cut -d',' -f1,2 >> PoD_students.csv   #cut: -d'' separatore // -f.. colonne
grep "Physics" LCP_22-23_students.csv | cut -d',' -f1,2 >> Physics_students.csv
```

1\.c For each letter of the alphabet, count the number of students whose surname starts with that letter. 

```python
lettere=( {A..Z} ) #array di tutte le lettere
echo "# PoD Students"
for i in "${lettere[@]}"; do # ${arr[@]} per accedere a tutti i valori e loopare
    echo "$i : $(grep -c "^$i" PoD_students.csv) "  #echo -n non va a capo // $^i per accedere alla variabile PRIMO CARATTERE
done

echo "# Physics Students"
for i in "${lettere[@]}"; do # ${arr[@]} per accedere a tutti i valori e loopare
    echo "$i : $(grep -c "^$i" Physics_students.csv) "  #echo -n non va a capo // $^i per accedere alla variabile
done
```

1\.d Find out which is the letter with most counts.

```python
lettere=( {A..Z} ) #array di tutte le lettere
most=0
letter=""

for i in $(seq 0 25); do 
    a=$(grep -c "^${lettere[$i]}" PoD_students.csv)
    if [ $a -gt $most ]; then 
        most=$a
        letter=${lettere[$i]}
    fi
done

echo "Letter with most count: $letter ,$most times"
#For PoD is B = 12
```

```python
lettere=( {A..Z} ) #array di tutte le lettere
most=0
letter=""

for i in $(seq 0 25); do 
    a=$(grep -c "^${lettere[$i]}" Physics_students.csv)
    if [ $a -gt $most ]; then 
        most=$a
        letter=${lettere[$i]}
    fi
done

echo "Letter with most count: $letter ,$most times."
#For Physics is F = 2
```

1\.e Assume an obvious numbering of the students in the file (first line is 1, second line is 2, etc.), group students "modulo 18", i.e. 1,19,37,.. 2,20,38,.. etc. and put each group in a separate file  

```python
touch LCP_22-23_students_num.csv

for (( i=2; i<=68; i++ )); do 
    echo "$((i-1)),$(sed -n "${i}p" LCP_22-23_students.csv)" >> LCP_22-23_students_num.csv  #-n suppress auto print of sed
done
```

```python
j=0
for (( i=1; i<=18; i++ )); do 
    rm "mod18_student_group_${i}".csv
    touch "mod18_student_group_${i}".csv
    counter=$i
    while [ $counter -lt 68 ]; do
    echo "$(grep "^${counter}," LCP_22-23_students_num.csv)" >> "mod18_student_group_${i}".csv
    let counter+=18
    done
done
```
