1\. Find the row, column and overall means for the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python

m =np.arange(12).reshape((3,4))
m.shape
````

````python
m = np.arange(12).reshape((3,4))
mean_row, mean_col = [], []
# row mean
for i in range(m.shape[0]):
    mean = m[i,:].mean()
    mean_row.append(mean)
# column mean
for i in range(m.shape[1]):
    mean = m[:,i].mean()
    mean_col.append(mean)
#overall mean
mean_ov = (m.flatten()).mean()

print('Mean on rows: \n', mean_row)
print('Mean on columns: \n', mean_col)
print('Overall Mean: \n', mean_ov)
````

2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
# Define vectors
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
````

````python
# Outer product using prebuilded function
print(np.outer(u, v, out=None))
````

````python
# Using for nested loop
outer = np.zeros((u.shape[0], v.shape[0]))
for i in range(u.shape[0]):
    for j in range(v.shape[0]):
        outer[i,j] = u[i]*v[j]
print(outer)
````

````python
# Using list comprehension
outer = [ u[i]*v[j] for i in range(u.shape[0]) for j in range(v.shape[0]) ]
outer = np.array(outer).reshape(4,4)
print(outer)
````

````python
# Using broadcasting operations
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

outer = np.array([ [u[i], u[i], u[i], u[i]] for i in range(u.shape[0]) ])

print(outer * v)
````

4\. Use np.linspace to create an array of 100 numbers between 0 and 2Ï€ (includsive).

  * Reverse the array using slice notation
  * Extract elements where the absolute difference between the sine and cosine functions evaluated at that element is less than 0.1
  * Make a plot showing the sin and cos functions and indicate where they are close

````python
pi = np.linspace(0, np.pi, 100)
view = pi[::10] # Estrae ogni 10 elementi -> start=0, stop=0 (un val), step=10
print("10th El. array: \n", view, "\n")
view_r = view[::-1] # revesed step
print("Reversed array: \n", view_r, "\n")

sin, cos = np.sin(pi), np.cos(pi) 
#print("sin[arr]: \n", sin)
#print("cos[arr]: \n", cos)

for i in range(pi.shape[0]):
    if (abs(sin[i] - cos[i]) < 0.1):
        print(f" |sin[x] - cos[x]| < 0.1 <-> x = {pi[i]} \n")

import matplotlib.pyplot as plt

plt.scatter(pi, sin, marker='.', color='r', label='sin_func')
plt.scatter(pi, cos, marker='.', color='b', label='cos_func')
plt.xlabel('pi')
plt.ylim([-1,1])
plt.xlim([0, np.pi])
plt.legend()
plt.tight_layout()
plt.show
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
np.set_printoptions(precision=1, suppress=True) #-> to fix np rounding
````

````python
pos = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])

route_66 = {"Chicago": 0, "Springfield": 1, "Saint-Louis": 2, "Tulsa": 3, "Oklahoma City": 4, "Amarillo": 5, "Santa Fe": 6, "Albuquerque": 7, "Flagstaff": 8, "Los Angeles": 9}

m = np.zeros((10,10))
for i in range(10):
    for j in range(10):
        m[i,j] = abs(pos[i] - pos[j])
print("Distances in miles: \n", m, "\n")

# 1 mile = 1.609344 km
m = m * 1.609344
print("Distances in km: \n", m, "\n")
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import time

def prime(N):
    prime = []
    
    num = np.arange(1,N+1,1)
    for i in range(1,10): # skip i = 0 <-> num[0] = 1 because all prime numbers must be mulpile of 1
        if (num[i] == 0):
            continue
        mask = (num % num[i] == 0)
        num[mask] = 0
    
    prime = np.array([num[i] for i in range(N) if num[i] != 0])
    
    return prime

for i in range(2,10):
    N = 10**i
    t0 = time.time()
    p = prime(N)
    t1 = time.time()
    print(f"Prime < {N}: \n", p, "\n")
    print(f"Time to compute:= {abs(t0 -t1)} s", "\n")
````
