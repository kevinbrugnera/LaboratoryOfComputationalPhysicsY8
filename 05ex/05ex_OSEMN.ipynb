2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
matrix = np.random.rand(5,5)
print(matrix)
```

```python
file_name = '/mnt/c/Users/kevin/Desktop/LaboratoryOfComputationalPhysicsY8/05ex/data.txt'
with open(file_name, mode='w') as f:
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            f.write(f"{str(matrix[i,j])} ") # spacebar to separate
        f.write("\n")

with open(file_name, mode="r") as f:
    for line in f:
        print(line)
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
file_name = '/mnt/c/Users/kevin/Desktop/LaboratoryOfComputationalPhysicsY8/05ex/data.txt'
file_name_csv = '/mnt/c/Users/kevin/Desktop/LaboratoryOfComputationalPhysicsY8/05ex/data.csv'
with open(file_name, mode="r") as txt:
    with open(file_name_csv, mode="w") as csv:
        for line in txt:
            csv.write(f"{line[:-2].replace(" ", ",")}\n") # replace until last 2 letters (" " + "\n") -> dont want line ending with comma
with open(file_name_csv, mode="r") as f:
    for line in f:
        print(line)
```
